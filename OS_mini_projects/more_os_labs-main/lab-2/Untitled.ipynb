```python
import matplotlib.pyplot as plt
```

```python
x = [64,128,256,512]
```

```python
y_mean_matrix_time = [12442.6,96087.7,1920828.3,17910933.2]
fig, ax = plt.subplots()
ax.plot(x, y_mean_matrix_time)
ax.set_yscale('log')
ax.set_title('exp scale prepaging matrix_size versus mean matrix mult time')
```

```python
y_mean_combine_time = [12735.2,96379.4,1921164.4,17911401.9]
fig, ax = plt.subplots()
ax.plot(x, y_mean_combine_time)
ax.set_yscale('log')
ax.set_title('exp scale prepaging matrix_size versus mean matrixmult + mmap time')
```

```python
y_mean_mmap_time_dmand = [123.3,106,110.1,119.5]
fig, ax = plt.subplots()
ax.plot(x, y_mean_mmap_time_dmand)
ax.set_yscale('linear')
ax.set_title('linear scale prepaging matrix_size versus mean mmap time for demand paging')
```
